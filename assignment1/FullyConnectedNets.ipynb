As a sanity check, run the following to check the initial loss and to gradient check the network both with and without regularization. This is a good way to see if the initial losses seem reasonable.

For gradient checking, you should expect to see errors around 1e-7 or less.

````python
np.random.seed(231)
N, D, H1, H2, C = 2, 15, 20, 30, 10
X = np.random.randn(N, D)
y = np.random.randint(C, size=(N,))

for reg in [0, 3.14]:
    print("Running check with reg = ", reg)
    model = FullyConnectedNet(
        [H1, H2],
        input_dim=D,
        num_classes=C,
        reg=reg,
        weight_scale=5e-2,
        dtype=np.float64
    )

    loss, grads = model.loss(X, y)
    print("Initial loss: ", loss)

    # Most of the errors should be on the order of e-7 or smaller.
    # NOTE: It is fine however to see an error for W2 on the order of e-5
    # for the check when reg = 0.0
    for name in sorted(grads):
        f = lambda _: model.loss(X, y)[0]
        grad_num = eval_numerical_gradient(f, model.params[name], verbose=False, h=1e-5)
        print(f"{name} relative error: {rel_error(grad_num, grads[name])}")
````

## Inline Question 2:

AdaGrad, like Adam, is a per-parameter optimization method that uses the following update rule:

```
cache += dw**2
w += - learning_rate * dw / (np.sqrt(cache) + eps)
```

John notices that when he was training a network with AdaGrad that the updates became very small, and that his network was learning slowly. Using your knowledge of the AdaGrad update rule, why do you think the updates would become very small? Would Adam have the same issue?


## Answer:

> Well, instead of *like Adam* as the question states, I would say *indeed like Adam* but more like *RMSProp*. 

AdaGrad accumulates the sum of squared gradients indefinitely, while RMSProp uses a decaying moving average. This means that in AdaGrad, the accumulated value in `cache` can grow very large over time, especially if the gradients are consistently non-zero. 
As a result, the denominator in the update rule becomes very large, leading to very small updates to the parameters `w`, which is why John observed slow learning. While in `RMSProp`, thanks to the decay factor, the `cache` value remains more responsive to recent gradients and gradually *forgets* older gradients, which creates a moving average that "forgets" old gradients. Recent gradients have more influence, so `cache` doesn't grow unboundedly.
Adam also uses a decaying moving average rather than an accumulating sum, like RMSProp, in my code implementation:
```python
config["v"] = config["beta2"] * config["v"] + (1 - config["beta2"]) * (dw ** 2)
```
the `beta2` serves the same purpose as the decay factor in RMSProp. Therefore, Adam would not likely face the same issue as AdaGrad, as it maintains a more balanced and adaptive learning rate for each parameter.
